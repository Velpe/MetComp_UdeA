# Interpolación vía Polinomios de Lagrange

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_list=[0,0.6]
y_list=[np.log(1+0.0),np.log(1+0.6)]

fig=plt.figure()
ax=plt.axes
plt.plot(x_list,y_list,"-")
plt.show()
```

```python
def PolLag(x_list,y_list,x):
    suma=0
    for i in range(0,len(x_list)):
        prod=1
        for j in range(0,len(x_list)):
            if(j!=i):
                prod=prod*(x-x_list[j])/(x_list[i]-x_list[j])
        suma=suma + y_list[i]*prod
    return(suma)
```

```python
y_derivado=PolLag(x_list,y_list,0.45)
y_exacto=np.log(1+0.45)
error=abs((y_derivado-y_exacto)/y_exacto * 100)
print()
```

```python
PolLag(x_list,y_list,3)
```

```python
x_array=np.linspace(min(x_list),max(x_list),100)
y_array=[]
```

```python
for i in range(0,len(x_array)):
    y_array.append(PolLag(x_list,y_list,x_array[i]))
```

```python
fig=plt.figure()
ax=plt.axes
plt.plot(x_array,y_array)
plt.plot(x_list,y_list,"*")
plt.show()
```
