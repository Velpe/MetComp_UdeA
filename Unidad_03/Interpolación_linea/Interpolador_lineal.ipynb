```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
datos=np.loadtxt("datos.dat")

    
#print(datos)
#print(x_list)
#print(y_list)
```

```python
x_list=[]
y_list=[]
for i in range(0,datos.shape[0]):
    x_list.append(datos[i][0])
    y_list.append(datos[i][1])
def Interpolador(x_list,y_list,x):
    for i in range(0,len(x_list)-1):
        if(x>=x_list[i] and x<=x_list[i+1]):
            x_prev=x_list[i] #x_i
            y_prev=y_list[i] #y_i
            x_post=x_list[i+1] #x_i+1
            y_post=y_list[i+1] #y_i+1
            
            y=((y_post-y_prev)/(x_post - x_prev))*(x-x_prev) + y_prev
            return y
x_interpol=np.linspace(min(x_list),max(x_list),100)
y_interpol=[]

for i in range(0,len(x_interpol)):
    y_interpol.append(Interpolador(x_list,y_list,x_interpol[i]))
print(Interpolador(x_list,y_list,0.5))
```

```python
fig=plt.figure()
ax=plt.axes()
ax.plot(x_list,y_list,"*")
ax.plot(x_interpol,y_interpol)
ax.plot(x_interpol,np.sin(x_interpol))
plt.show()
```
